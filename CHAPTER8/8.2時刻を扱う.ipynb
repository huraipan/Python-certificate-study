# **8.2.1 時刻を取得する**

```python
# 時刻を取得する
import time

print(time.gmtime())
print(time.localtime())
print(time.strftime('%Y-%m-%d', time.localtime()))
print(time.time())
```

# **8.2.2 時刻オブジェクトーstruct_time**

```python
# struct_timeのサンプルコード
local = time.localtime()
utc = time.gmtime()
print(local.tm_zone, local.tm_gmtoff)
print(utc.tm_zone, utc.tm_gmtoff)

print(local.tm_hour)
print(utc.tm_hour)
```

# **8.2.3 スレッドの一時停止ーsleep()**

```python
# sleepのサンプルコード
for i in range(5):
    time.time()
    time.sleep(0.5)
```

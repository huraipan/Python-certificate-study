# **8.1.1 日付を扱う-dateオブジェクト**

```python
# dateのサンプルコード
from datetime import date

ganjitsu = date(2021, 1, 1)
print(ganjitsu)
print(ganjitsu.year, ganjitsu.month, ganjitsu.day)
print(ganjitsu.weekday())
print(ganjitsu.isoformat())
print(date.fromisoformat('2021-01-01'))
print(str(ganjitsu))
print(ganjitsu.strftime('%Y/%m/%d'))
print(ganjitsu.strftime('%Y %m %d (%a)'))
print(date.today())
today = date.today()
print(f'今日は{today:%Y年%m月%d日}')
```

# **8.1.2 時刻を扱う-timeオブジェクト**

```python
# timeのサンプルコード
from datetime import time

print(time())
print(time(16, 12, 25))
print(time(minute=10))
print(time(second=10))
print(time(microsecond=10))

now = time(16, 12, 25)
print(now.hour, now.minute, now.second, now.microsecond)
print(now.isoformat())
print(now.fromisoformat('16:12:25'))
print(now.fromisoformat('16:12:25.000384'))
print(now.fromisoformat('16:12:25+04:00'))
print(now.strftime('%H:%M'))
print(str(now))
```

# **8.1.3 日時を扱うーdatetimeオブジェクト**

```python
# datetimeのサンプルコード
from datetime import datetime
from zoneinfo import ZoneInfo

now = datetime.now()
print(now.date())
print(now.time())
print(now.isoformat())

print(datetime.fromisoformat('2021-02-14'))
print(datetime.fromisoformat('2021-02-14T00:05:23'))
print(datetime.fromisoformat('2021-02-14T17:26:08.585404'))

print(now.strftime('%Y/%m/%d'))
print(datetime.strptime('2021/02/14', '%Y/%m/%d'))

print(datetime(2021, 5, 24, 22, 50, 0, 0, tzinfo=ZoneInfo('Asia/Tokyo')))
```

# **8.1.4 日時の差を扱うーtimedeltaオブジェクト**

```python
# timedeltaのサンプルコード
from datetime import date, datetime, time, timedelta

today = date.today()
print(today)

newyearsday =date(2025, 1, 1)
print(newyearsday - today)

week = timedelta(days = 7)
print(today + week)
print(today + week * 2)
print(today - week)

print(datetime(2021, 5, 24, 21, 0, 0, 0) - datetime(2021, 5, 24, 20, 0, 0, 0))
```

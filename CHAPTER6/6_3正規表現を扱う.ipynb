# **6.3.1 基本的な関数ーsearch、match**

```python
# 基本的な正規表現のマッチング処理
import re

print(re.search('a.c', 'abcde'))
print(re.match('a.c', 'abcde'))
print(re.search('.c', 'abcde'))
print(re.match('.c', 'abcde'))
```

# **6.3.2 reモジュールの定数**

```python
# フラグのサンプルコード
print(re.search('\w', 'あいうえおABC'))
print(re.search('\w', 'あいうえおABC', flags=re.A))

print(re.search('[abc]+', 'ABC'))
print(re.search('[abc]+', 'ABC', re.I))

print(re.search('a.c', 'a\nc'))
print(re.search('a.c', 'a\nc', re.S))
print(re.search('a.c', 'A\nc', re.I | re.S))
```

# **6.3.3 正規表現オブジェクト**

```python
# 関数と正規表現オブジェクト
pattern = re.compile('a.c')
print(pattern.search('abced'))
print(re.search('a.c', 'abcde'))
```

```python
# マッチオブジェクトを取得するメソッドのサンプルコード
regex = re.compile('[a-n]+')
print(type(regex))

print(regex.search('python'))
print(regex.match('python'))
print(regex.fullmatch('eggs'))
print(regex.fullmatch('egg'))
```

```python
# 正規表現パターンで分割、置換するサンプルコード
regex2 = re.compile('[-+()]')
print(regex2.split('080-1234-5678'))
print(regex2.split('(080)1234-5678'))
print(regex2.split('+81-80-1234-5678'))
print(regex2.sub('', '+81-80-1234-5678'))
```

```python
# 正規表現パターンにマッチする結果をすべて取得するサンプルコード
regex3 = re.compile('\d+')
print(regex3.findall('080-1234-5678'))

i = regex3.finditer('+81-80-1234-5678')
print(type(i))

for m in i:
    print(m)
```

# **6.3.4 マッチオブジェクト**

```python
# group()と[]のサンプルコード
m = re.match(r'(\d+)-(\d+)-(\d+)', '080-1234-5678')
print(m.group(0))
print(m.group(1))
print(m.group(2))
print(m.group(1, 2, 3))
print(m[0])
print(m[1])
```

```python
# 名前付きサブグループを指定するサンプルコード
m2 = re.match(r'(?P<last>\w+) (?P<first>\w+)', '鈴木 たかのり')

print(m2.group('last'))
print(m2.group('first'))
print(m2.group('last', 'first'))

print(m2[1])
print(m2[2])
```

```python
# マッチオブジェクトのメソッドのサンプルコード
m = re.match(r'(\w+) ?(\w+)? (\w+)', 'Guido van Rossum')
print(m.groups())

m = re.match(r'(\w+) ?(\w+)? (\w+)', 'Takanori Suzuki')
print(m.groups())
print(m.groups('van'))

m = re.match(r'(?P<last>\w+) (?P<first>\w+)', '鈴木 たかのり')
print(m.groupdict())
print(m.expand(r'名前: \1\2'))
print(m.expand(r'名字(\g<last>) 名前(\g<first>)'))
```

# **6.4.1 Unicode文字と文字の名前を変換する**

```python
# unicodedataのサンプルコード
import unicodedata

print(unicodedata.lookup('Latin Small Letter a'))
print(unicodedata.lookup('BEER MUG'))
# print(unicodedata.lookup('UNKNOWN CHARACTER'))    KeyError

for chr in ('A', 'Ａ', '1', 'い', 'ｶ', '㌧', '☃', '🍣'):
    print(unicodedata.name(chr))
```

# **6.4.2 Unicode文字列の正規化**

```python
print(unicodedata.normalize('NFC', 'ｱアAＡ@＠！!'))
print(unicodedata.normalize('NFKC', 'ｱアAＡ@＠！!'))
print(unicodedata.normalize('NFKC', '㌧'))
```

```python
# NFKCとNFKDの結果の違いを確認する
nfkc = unicodedata.normalize('NFKC', 'ガ')
nfkd = unicodedata.normalize('NFKD', 'ガ')

print(nfkc, nfkd)

print(nfkc.encode('utf-8'))
print(nfkd.encode('utf-8'))

nfkd_encoded = nfkd.encode('utf-8')
print(nfkd_encoded.decode('utf-8'))
print(nfkd_encoded[:3].decode('utf-8'))
print(nfkd_encoded[3:].decode('utf-8'))
```

# **10.2.1 インメモリなテキストストリームを扱うーStringIO**

```python
# StringIOクラスの基本的な使い方
import io

stream = io.StringIO("this is test\n")
print(stream.read(10))
print(stream.tell())
print(stream.seek(0, io.SEEK_END))
print(stream.write('test'))
print(stream.getvalue())

stream.close()
# stream.write('test')      閉じたあとに書き込もうとすると例外を送出する

with io.StringIO() as stream:
    print(stream.write('test'))

the_zen_of_python = """The Zen of Python, by Tim Peters
Beauriful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
"""

stream = io.StringIO(the_zen_of_python)
for line in stream:
    print(line)
```

# **10.2.2 インメモリなバイナリストリームを扱うーBytesIO**

```python
# ByesIOクラスの基本的な使い方
stream = io.BytesIO(b'abcdefg')
print(stream.read(5))
print(stream.tell())
print(stream.seek(0, io.SEEK_END))
print(stream.write(b'test'))
print(stream.getvalue())
stream.close()
# stream.wirte(b'test')     閉じたあとに書き込もうとすると例外を送出する

with io.BytesIO() as stream:
    print(stream.write(b'test'))

stream = io.BytesIO(b'abcdefg')
view = stream.getbuffer()
view[2:4] = b'56'
print(stream.getvalue())
```

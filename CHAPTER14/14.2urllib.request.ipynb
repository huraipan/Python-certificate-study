# **14.2.1 指定のURLを開くーurlopen()**

```python
# GETメソッドを使う
from urllib import request

with request.urlopen('https://httpbin.org/get') as f:
    res = f.read()[:92]

print(res)
```

```python
# ファイルのダウンロード
file_data = request.urlopen('https://httpbin.org/image/jpeg').read()
with open('./test.jpg', 'wb') as f:
    print(f.write(file_data))
```

```python
# POSTメソッドを使う
data = 'key1=value1&key2=value2'
res = request.urlopen('https://httpbin.org/post', data=data.encode())
print(res.status)
```

# **14.2.2 GET、POST以外のHTTPメソッドを扱う**

```python
# DELETEメソッドを使うサンプル
data = 'key1=value1&key2=value2'
req = request.Request('https://httpbin.org/delete', data=data.encode(), method='DELETE')
with request.urlopen(req) as f:
    res_body = f.read()[:110]
    res_status = f.status

print(res_status)
print(res_body)
```

```python
# カスタムヘッダー
headers = {'Accept': 'application/json'}
request.Request('https://httpbin.org/get', headers=headers)
```

# **14.2.3 レスポンスモジュール**

```python
# レスポンスオブジェクトの取得
res = request.urlopen('https://httpbin.org/get')
print(res.url)
print(res.status)
print(res.headers)
```

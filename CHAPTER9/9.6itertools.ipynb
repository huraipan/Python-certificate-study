# **9.6.1 イテラブルオブジェクトを連結するーchain()**

```python
# itertools.chain()のサンプルコード
import itertools

it = itertools.chain(['A', 'B'], 'ab', range(3))
for element in it:
    print(element)
```

# **9.6.2 連続する値をーgroupby()**

```python
# itertools.groupby()のサンプルコード
for value, group in itertools.groupby('aaabbcdddaabb'):
    print(f'{value}: {list(group)}')
```

```python
# 入力データをソートしてからgroupby()を実行する
sorted_text = ''.join(sorted('aaabbcdddaabb'))
print(sorted_text)

for value, group in itertools.groupby('aaabbcdddaabb'):
    print(f'{value}: {list(group)}')
```

```python
# keyに関数を指定したサンプルコード
def is_odd(num):
    return num % 2 == 1

numbers = [10, 20, 31, 11, 3, 4]
for value, group in itertools.groupby(numbers,is_odd):
    print(f'{value}: {list(group)}')
```

# **9.6.3 イテレーターから範囲を指定してい値を取得するーislice()**

```python
# itertools.islice()のサンプルコード
li = list(range(10))
print(li)

islice_object = itertools.islice(li, 5)
print(islice_object)
print(list(islice_object))
```

# **9.6.4 複数のイテラブルオブジェクトの要素からタプルを作成するーzip(),zip_longest()**

```python
# zip()関数のサンプルコード
it1 = (1, 2, 3)
it2 = ('a', 'b', 'c')
it3 = 'あいう'

for v in zip(it1, it2, it3):
    print(v)
```

```python
# 長さの異なるイテラブルオブジェクトを指定する
it1 = (1, 2, 3, 4, 5)
it2 = ['abc', 'ABC', '123']

for v in zip(it1, it2):
    print(v)
```

```python
# itertools.zip_longest()のサンプルコード
it1 = (1, 2, 3, 4, 5)
it2 = ['abc', 'ABC', '123']

for v in itertools.zip_longest(it1, it2):
    print(v)
```

```python
# fillvalue引数を指定する
for v in itertools.zip_longest('abcde', '123', 'あいうえ', fillvalue='-'):
    print(v)
```

# **9.6.5 データを組み合わせたイテレーターを取得する**

```python
# iertools.product()のサンプル
import itertools

print(list(itertools.product('ABC', [1, 2, 3])))

print([r[0] + r[1] for r in itertools.product('ABC', repeat=2)])

results = []
for i in 'ABC':
    for j in 'ABC':
        results.append(i+j)
```

```python
# itertools.permutations()のサンプル
print(list(itertools.permutations('ABC')))
results = itertools.permutations('ABCD', 2)
print([r[0] + r[1] for r in results])
```

```python
# itertools.combinations()のサンプルコード
print(list(itertools.combinations('ABC', 2)))
results = itertools.combinations('ABCD', 2)
print([r[0] + r[1] for r in results])
results = itertools.permutations('ABCD', 2)
print([r[0] + r[1] for r in results])
```

```python
# itertools.combinations_with_replacement()のサンプル
print(list(itertools.combinations_with_replacement('ABC', r=2)))
print(list(itertools.combinations('ABC', r=2)))
```

# **9.4.1 定数値を定義する**

```python
# enumのサンプル
import enum

class Nengo(enum.Enum):
    SHOWA = 1
    HEISEI = 2
    REIWA = 3
```

```python
# enum.auto()を使って値の指定を省略する
class Nengo(enum.Enum):
    SHOWA = enum.auto()
    HEISEI = enum.auto()
    REIWA = enum.auto()
```

```python
# uniqueデコレーター
@enum.unique
class Spam(enum.Enum):
    HAM = 1
    EGG =1
```

# **9.4.2 定数を呼び出す**

```python
# 定数を呼び出す方法
print(Nengo.REIWA)
print(Nengo['REIWA'])
print(Nengo(3))
```

```python
# 定数の名前と値を取得する
showa = Nengo.SHOWA
print(showa.name)
print(showa.value)
```

```python
# 定数のイテレーター
class Spam(enum.Enum):
    HAM = 1
    EGG = 2
    BACON = 1

list(Spam)
```

# **9.4.3 定数同士を比較する**

```python
# 定数の比較
class Spam(enum.Enum):
    HAM = 1
    EGG = 2
    BACON = 2

print(isinstance(Spam.HAM, Spam))
print(Spam.HAM == Spam.HAM)
print(Spam.HAM == Spam.EGG)
print(Spam.EGG == Spam.BACON)
```

```python
# 異なる型の比較
class OtherSpam(enum.Enum):
    HAM = 1
    EGG = 2
    BACON = 2

print(Spam.HAM == OtherSpam.HAM)
print(Spam.HAM == 1)
```

# **9.7.1 浅いコビーを行うーcopy()関数**

```python
# 変数の代入による参照の作成
values = ['a', 'b', 'c', 'd']
v_ref = values
print(v_ref)

v_ref[1] = 'e'
print(v_ref)
print(values)
```

```python
# copy()関数による浅いコピー
import copy

values = ['a', 'b', 'c', 'd']
val_cp = copy.copy(values)
print(val_cp)
val_cp[1] = 'e'
print(val_cp)
print(values)
```

```python
# copy()関数で子オブジェクトが参照になる例
values = [[0, 1], [2, 3], [4, 5]]
val_cp = copy.copy(values)
val_cp.append([6,7])
print(val_cp)
print(values)
val_cp[1][0] = 8
print(val_cp)
print(values)
```

# **9.7.2 深いコピーを行うーdeepcopy()関数**

```python
# deepcopy()関数の例
values = [[0, 1], [2, 3], [4, 5]]
val_cp = copy.deepcopy(values)
val_cp[1][0] = 8
print(val_cp)
print(values)
```

```python
# id()関数を使用したオブジェクト識別値の確認
values = [[0, 1], [2, 3], [4, 5]]
val_cp = copy.copy(values)
val_deep_cp = copy.deepcopy(values)
print(id(values))
print(id(val_cp))
print(id(val_deep_cp))

print(id(values[0]))
print(id(val_cp[0]))
print(id(val_deep_cp[0]))
```

# **9.7.3 copy:よくある使い方**

```python
# ループ処理中に辞書オブジェクトの編集を行う
authors = ['ryu22e', 'sugita', 'fukuda']
attrib = {'attribute': 'author'}
book_authors = []
for name in authors:
    cp_attrib = copy.copy(attrib)
    cp_attrib['name'] = name
    book_authors.append(cp_attrib)

book_authors
```

```python
# copyモジュールを使用したインスタンスのコピー
class Author:
    def __init__(self, name, age):
        self.name = name
        self.age = age

author1 = Author('kadowaki', 25)
author2 = Author('terada', 20)
author3 = copy.deepcopy(author1)
author3.name = 'takanory'
print(author1.name, author1.age)
print(author2.name, author2.age)
print(author3.name, author3.age)
```

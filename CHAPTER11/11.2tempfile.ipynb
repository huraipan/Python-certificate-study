# **11.2.1 一時ファイルを作成する**

```python
# 一時ファイルの使用例
import tempfile

with tempfile.TemporaryFile() as tmpf:
    print(tmpf.write(b'test test test\n'))
    print(tmpf.seek(0))
    print(tmpf.read())

tmpf = tempfile.TemporaryFile()
print(tmpf.write(b'Hello tempfile\n'))
print(tmpf.seek(0))
print(tmpf.read())
tmpf.close()
```

```python
# 名前を持った一時ファイルの作成
from pathlib import Path

tmpf = tempfile.NamedTemporaryFile()
print(tmpf.name)

p = Path(tmpf.name)
print(p.exists())

tmpf.close()
print(p.exists())
```

# **11.2.2 一時ディレクトリを作成する**

```python
# 一時ディレクトリの使用例
from pathlib import Path

with tempfile.TemporaryDirectory() as tmpdirname:
    tmpdirname
    p = Path(tmpdirname)
    print(p.exists())
    p2 = p / 'hoge.txt'
    p2.touch()
    print(p2.exists())

print(p.exists())
print(p2.exists())
```

# **11.3.1 ファイルをコピーする**

```python
# ファイルをコピーする
import shutil

shutil.copyfile('a.txt', 'b.txt')
shutil.copy('a.txt', 'c.txt')
shutil.copy2('a.txt', 'd.txt')
```

```python
# ファイルをディレクトリにコピーする
shutil.copy('a.txt', 'e/')
shutil.copy2('a.txt', 'e/')
# shutil.copyfile('a.txt', 'e/')        IsDirectoryError
```

# **11.3.2 再帰的にディレクトリやファイルを操作する**

```python
# 指定ディレクトリをコピーするサンプル
ignore = shutil.ignore_patterns('*.pyc', '*.swp')
print(ignore)
print(shutil.copytree('./from', './to', ignore=ignore))
```
